Task 2: You will be working on Names corpus. Import it and do the following:

```python
from nltk.corpus import names 
from nltk import FreqDist, ConditionalFreqDist

nltk.download('names')
```

2.1. Define a conditional frequency distribution over the Names Corpus that allows you to
see which last letters are more frequent for males versus females. 

```python
cfd = nltk.ConditionalFreqDist(
    (fileid, name[1]) 
    for fileid in names.fileids() 
    for name in names.words(fileid))

cfd.plot()
```

2.2.Plot the conditional frequency distribution of the number of female and male names
ending with each letter of the alphabet.

```python
cfd = nltk.ConditionalFreqDist(
    (fileid, name[-1]) 
    for fileid in names.fileids() 
    for name in names.words(fileid))

cfd.plot()
```
